```python
def region_function(x):
    if x['region'] == 'Southeast AL AR FL GA KY LA MS NC SC TN VA WV':
        return 'Southeast'
    elif x['region'] == 'Far West AK CA HI NV OR WA':
        return 'Farwest'
    elif x['region'] == 'Rocky Mountains CO ID MT UT WY':
        return 'Mountains'
    elif x['region'] == 'New England CT ME MA NH RI VT':
        return 'New England'
    elif x['region'] == 'Mid East DE DC MD NJ NY PA':
        return 'Mideast'
    elif x['region'] == 'Great Lakes IL IN MI OH WI':
        return 'Great Lakes'
    elif x['region'] == 'Plains IA KS MN MO NE ND SD':
        return 'Plains'
    elif x['region'] == 'US Service schools':
        return 'US Services'
    else:
        return None
```

```python
df2['clear_region'] = df2.apply(region_function, axis = 1)
```

```python
df2['clear_region'].head()
```

```python
df2.clear_region.value_counts()
```

```python
plt.figure(figsize = [10, 10])
barplot_regions = sns.countplot(x = 'clear_region', data = df2)
for p in barplot_regions.patches:
    barplot_regions.annotate(format(p.get_height(), '.1f'), 
                        (p.get_x()+p.get_width()/2., p.get_height()), 
                         ha = 'center', va='center', 
                         xytext=(0, 5),
                         textcoords = 'offset points')
plt.xlabel("Region")
plt.ylabel('Count')
plt.title("Number of Universities, grouped by Regions")
```

```python
df2.groupby('clear_region').Controlofinstitution.value_counts()
```

```python
plt.figure(figsize = [10, 10])
region_sector_barplot = sns.countplot(x = 'clear_region', hue = 'Controlofinstitution', data = df2)
for p in region_sector_barplot.patches:
    region_sector_barplot.annotate(format(p.get_height(), '.1f'), 
                                  (p.get_x()+p.get_width()/2., p.get_height()),
                                  ha = 'center', va = 'center', 
                                  xytext = (0, 5), 
                                  textcoords = 'offset points')
plt.xlabel("Region")
plt.ylabel('Count')
plt.title("Number of Universities by Sector, grouped by Regions")
plt.legend(loc = (0.7, 0.9))
```

```python
plt.figure(figsize = [9, 9])
sns.boxplot(x = 'clear_region', y = 'Tuitionandfees201314', data = df2)
plt.xlabel('Region')
plt.ylabel('Total Tuition and Fees, 2013-2014')
plt.title('Distrbution of Total Tuition and Fees by Regions')
```
